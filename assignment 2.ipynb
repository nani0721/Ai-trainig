```python
class Node:
    def __init__(self, name):
        self.name = name
        self.left = None  
        self.right = None 
def level_order_traversal(root):
    if not root:
        return
    queue = [root]
    while queue:
        current = queue.pop(0)
        print(current.name)
        if current.left:
            queue.append(current.left)
        if current.right:
            queue.append(current.right)
def in_order_traversal(root):
    if root:
        in_order_traversal(root.left)
        print(root.name)
        in_order_traversal(root.right)
ceo = Node("CEO")
manager1 = Node("Manager1")
manager2 = Node("Manager2")
employee1 = Node("Employee1")
employee2 = Node("Employee2")
ceo.left = manager1
ceo.right = manager2
manager1.left = employee1
manager2.right = employee2
print("Level-order Traversal (for email broadcast):")
level_order_traversal(ceo)

print("\nIn-order Traversal (for org chart layout):")
in_order_traversal(ceo)
```

```python
class Match:
    def __init__(self, name, player1=None, player2=None, left=None, right=None):
        self.name = name
        self.player1 = player1
        self.player2 = player2
        self.left = left
        self.right = right
        self.winner = None

    def play(self, winner):
        if winner not in [self.player1, self.player2]:
            raise ValueError(f"Winner must be {self.player1} or {self.player2}")
        self.winner = winner
        print(f"{self.name} winner: {winner}")

    def path_to_final(self, player, path=None):
        if path is None:
            path = []
        if self.winner == player:
            path.append(self.name)
            if self.left:
                self.left.path_to_final(player, path)
            if self.right:
                self.right.path_to_final(player, path)
        return path[::-1]


# Create quarter-final matches
Q1 = Match("Q1", "A", "B")
Q2 = Match("Q2", "C", "D")
Q3 = Match("Q3", "E", "F")
Q4 = Match("Q4", "G", "H")

Q1.play("A")
Q2.play("C")
Q3.play("F")
Q4.play("H")

# Create semi-finals using winners
S1 = Match("Semi1", Q1.winner, Q2.winner, Q1, Q2)
S2 = Match("Semi2", Q3.winner, Q4.winner, Q3, Q4)

S1.play("A")
S2.play("F")

# Final
Final = Match("Final", S1.winner, S2.winner, S1, S2)
Final.play("A")

# Output results
print("\nTournament Winner:", Final.winner)
print("Path to final for A:", " → ".join(Final.path_to_final("A")))



```

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def evaluate(node):
    if not node.left and not node.right:
        return int(node.value)
    l = evaluate(node.left)
    r = evaluate(node.right)
    if node.value == '+': return l + r
    if node.value == '-': return l - r
    if node.value == '*': return l * r
    if node.value == '/': return l / r

def to_inorder(node):
    if not node.left and not node.right:
        return node.value
    return f"({to_inorder(node.left)} {node.value} {to_inorder(node.right)})"

# Build expression: (3 + 5) * (2 - 1)
root = Node('*')
root.left = Node('+')
root.right = Node('-')
root.left.left = Node('3')
root.left.right = Node('5')
root.right.left = Node('2')
root.right.right = Node('1')

print("Expression:", to_inorder(root))
print("Result:", evaluate(root))


```
